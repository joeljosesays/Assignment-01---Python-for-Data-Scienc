```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(8,6))
sns.scatterplot(x=data['SL'],y=data['SW'],hue=data['Classification'])
plt.title('Relationship between SL and SW')
plt.xticks()
plt.yticks()
plt.show()

```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=data['PL'],y=data['PW'],hue=data['Classification'])
plt.title('Relationship between PL and PW')
plt.xticks()
plt.yticks()
plt.legend(loc=(1.02,0))
plt.show()
```

```python
data[['SL','PL']].plot.hist()
plt.title('Distribution of SL and PL')
plt.show()
```
